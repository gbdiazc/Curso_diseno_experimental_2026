## Funciones

Las **funciones** son uno de los conceptos fundamentales en el estudio del cálculo. En términos generales, una función es una regla que asigna a cada elemento de un conjunto llamado **dominio** un único elemento de otro conjunto denominado **codominio** o **recorrido**. Las funciones permiten modelar fenómenos del mundo real, como la trayectoria de un objeto, la variación de temperatura o el crecimiento de una población.

Una función puede representarse de diversas formas:

- Mediante una **expresión algebraica** (por ejemplo, $f(x) = x^2 + 1$),
- A través de una **tabla de valores**,
- Mediante una **descripción verbal**,
- O utilizando su **representación gráfica**.

### Definición formal

Dados dos conjuntos $X$ y $Y$, una **función** $f$ es una regla que asigna a cada elemento $x \in X$ un único elemento $y \in Y$, denotado por $f(x)$.

- El conjunto $X$ se denomina **dominio** de la función.  
- El conjunto de todos los valores $f(x)$ que resultan al aplicar la función a los elementos de $X$ se llama **codominio**, **recorrido** o **imagen** de la función.

Las funciones se suelen denotar con letras como $f$, $g$ o $h$. Si $f$ es una función y $x$ pertenece a su dominio, entonces $f(x)$ representa el **valor de la función en $x$**, y se lee “f de x”.

Una forma intuitiva de entender una función es pensarla como una **máquina**:

- Se introduce un valor $x$ (entrada),  
- Se aplica la regla definida por la función,  
- Se obtiene una salida $f(x)$.

Así, el **dominio** contiene las entradas y el **codominio** contiene todas las salidas posibles.


````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Crear figura y ejes con mayor margen horizontal
fig, ax = plt.subplots(figsize=(12, 7))

# Colores elegantes
dominio_color = '#4A90E2'       # Azul elegante
codominio_color = '#50E3C2'     # Verde menta
texto_color = '#222222'         # Gris oscuro para mejor contraste

# Dibujar conjuntos X (dominio) y Y (codominio)
ax.add_patch(patches.Circle((3, 3.5), 2, edgecolor=dominio_color, facecolor=dominio_color + '33'))
ax.add_patch(patches.Circle((9, 3.5), 2, edgecolor=codominio_color, facecolor=codominio_color + '33'))

# Elementos del dominio
dominio = [(2.7, 5), (2.5, 3.5), (3.3, 2.2)]
dom_texts = ['x₁', 'x₂', 'x₃']

# Elementos del codominio
codominio = [(8.7, 5), (9.5, 3.5), (8.5, 2.2)]
cod_texts = ['f(x₁)', 'f(x₂)', 'f(x₃)']

# Dibujar puntos, etiquetas y flechas
for i, ((dx, dy), (cx, cy)) in enumerate(zip(dominio, codominio)):
    ax.plot(dx, dy, 'o', color=dominio_color, markersize=12)
    ax.text(dx - 0.4, dy, dom_texts[i], fontsize=20, color=texto_color, weight='bold')
    ax.plot(cx, cy, 'o', color=codominio_color, markersize=12)
    ax.text(cx + 0.3, cy, cod_texts[i], fontsize=20, color=texto_color, weight='bold')
    ax.annotate('', xy=(cx, cy), xytext=(dx, dy),
                arrowprops=dict(arrowstyle='->', color='gray', lw=2))

# Etiquetas de conjuntos
ax.text(3, 6, 'Dominio (X)', fontsize=22, ha='center', color=dominio_color, weight='bold')
ax.text(9, 6, 'Codominio (Y)', fontsize=22, ha='center', color=codominio_color, weight='bold')

# Estética general
ax.set_xlim(0, 12)
ax.set_ylim(0.5, 6.5)
ax.axis('off')
plt.title('Relación entre Dominio y Codominio', fontsize=24, color=texto_color, pad=20)
plt.tight_layout()
plt.show()
````

### Representación gráfica

La forma más común de visualizar una función es a través de su **gráfica** en el plano cartesiano. Si una función $f$ tiene dominio $D$, su gráfica es el conjunto de pares ordenados:

$$
\{(x, f(x)) \mid x \in D\}
$$

Cada punto $(x, f(x))$ representa una relación entrada-salida: $x$ es la entrada y $f(x)$ es la salida. Estos puntos forman una curva o línea que describe el comportamiento de la función.

La gráfica de una función es útil para identificar visualmente propiedades como:

- Crecimiento o decrecimiento,
- Máximos o mínimos,
- Intersecciones con los ejes,
- Continuidad o discontinuidades.


````python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return np.sin(x) + 0.5 * x

# Valores para graficar
x_vals = np.linspace(0, 6, 400)
y_vals = f(x_vals)

# Crear figura
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la curva
ax.plot(x_vals, y_vals, color='deepskyblue', linewidth=2)

# Dibujar líneas verticales
x_points = [1, 2, 4]
for x in x_points:
    ax.vlines(x, 0, f(x), color='red', linestyle='-', linewidth=2)
    ax.text(x+0.2, f(x)/2, f'$f({x})$', fontsize=14, ha='center', va='bottom')

# Punto (x, f(x))
x = 4
ax.plot(x, f(x), 'o', color='blue')
ax.text(x-0.3, f(x)+0.2, r'$(x, f(x))$', fontsize=14)

# Anotar crecimiento
ax.annotate('Decrecimiento', xy=(2.5, f(2.5)), xytext=(1.7, f(1)+1),
            arrowprops=dict(facecolor='green', arrowstyle='->'),
            fontsize=14, color='green')

# Anotar decrecimiento
ax.annotate('Crecimiento', xy=(5, f(5)), xytext=(4.8, f(5)-0.4),
            arrowprops=dict(facecolor='green', arrowstyle='->'),
            fontsize=14, color='green')

# Ejes
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Etiquetas de ejes
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel('y', fontsize=16)

plt.tight_layout()
plt.show()
````

## Límites

Uno de los conceptos más importantes dentro del Cálculo es el concepto de *Límite*. Se dice que una función $f$ tiende hacia el *Límite* $l$ cerca de $a$, si se puede hacer que $f(x)$ este tan próxima como queramos de $l$, haciendo que $x$ esté suficientemente cerca de $a$, pero siendo distinta de $a$. Así por ejemplo si analizamos la función $f(x) = x^2 - x + 2$, para los valores cercanos a 2, podríamos ver los siguientes resultados.

````python
def f(x):
    return x**2 - x + 2

x = np.array([1, 1.5, 1.9, 1.95, 1.99, 1.999, 2.001, 2.05, 2.1, 2.2, 2.5, 3 ])
y = f(x)
tabla = pd.DataFrame({'x': x, 'f(x)': y})
tabla
````

de acuerdo con esta tabla, podemos ver que a medida que hacemos al valor de $x$ cercano a 2, vemos que $f(x)$ se hace muy cercana a 4. Incluso podríamos hacer a $f(x)$ tan cercana como queramos a 4, haciendo que $x$ este lo suficientemente cerca de 2. Por lo tanto, podemos expresar esta propiedad diciendo que el *"[Límite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico) de la función $f(x) = x^2 - x + 2$ cuando $x$ se acerca a 2 es igual a 4."* y lo podemos representar con la siguiente notación:

$$\lim_{x\to 2} \left(x^2 -x + 2\right) = 4$$

Gráficamente lo podemos ver del siguiente modo.

````python
plt.figure()
colors = ["#4B0000",
    "#5A0D0D",
    "#8B0000",
    "#9B111E",
    "#A50021",
    "#DC143C",
    "#FF0000",
    "#FA5B5A",
    "#FF6F61",
    "#FFA07A"]
x = np.array([1, 1.5, 1.9, 1.95, 1.99, 1.999, 2.001, 2.05, 2.1, 2.2, 2.5, 3 ])
y = f(x)
plt.plot(x, f(x))
x = [1.75,1.80,1.90,1.95,1.97,1.98,1.99]
print(x)

for i in range(len(x)):
    plt.scatter(x[i],f(x[i]),color = colors[i-1],label="x="+str(x[i]))
plt.scatter(2, 4, label="lim x-> 2", color=colors[-1])
plt.xlim([1.75,2.25])
plt.legend()
plt.title(r"Grafico de $f(x)=x^2 -x + 2$")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
````

### Las leyes de los límites

Calcular el valor exacto de los *Límites* muchas veces no suele tan fácil como reemplazar el valor de $a$ en $f(x)$. Es por esto que es importante conocer algunas propiedades de los *Límites*, ellas son:

1. **Ley de la suma:** El límite de la suma de dos funciones es la suma de sus límites.
2. **Ley de la diferencia:** El límite de la diferencia de dos funciones es la diferencia de sus límites.
3. **Ley del producto:** El límite del producto de dos funciones es el producto de sus límites.
4. **ley del múltiplo constante:** El límite de una constante por una función es la constante por el límite de la función.
5. **Ley del cociente:** El límite del cociente de dos funciones es el cociente de sus límites, siempre que el límite del denominador sea diferente de *cero*.

Es decir que si tenemos a la constante $C$ y a los límites $\lim_{x\to a} f(x)$ y $\lim_{x\to a} g(x)$. Entonces podemos expresar estas propiedades matemáticamente de la siguiente forma:

1- **Ley de la suma:** $\lim_{x\to a} [f(x) + g(x)] = \lim_{x\to a} f(x) + \lim_{x\to a} g(x)$.


2- **Ley de la diferencia:** $\lim_{x\to a} [f(x) - g(x)] = \lim_{x\to a} f(x) - \lim_{x\to a} g(x)$.


3- **Ley del producto:** $\lim_{x\to a} [f(x) \cdot g(x)] = \lim_{x\to a} f(x) \cdot \lim_{x\to a} g(x)$.


4- **ley del multiplo constante:** $\lim_{x\to a} [C \cdot f(x)] = C \cdot \lim_{x\to a} f(x)$.


5- **Ley del cociente:** $\lim_{x\to a} \left[\frac{f(x)}{g(x)}\right] = \frac{\lim_{x\to a} f(x)}{\lim_{x\to a} g(x)}$, si $\lim_{x\to a} g(x) \ne 0$.


### Calculando Límites con Python

Con [Python](https://www.python.org/), podemos resolver *[Límites](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico)* fácilmente utilizando la librería [SymPy](http://www.sympy.org/es/), la cual nos proporciona el objeto `Limit` para representarlos en [Python](https://www.python.org/). Su sintaxis es la siguiente: `Limit(función, variable, punto)`. Entonces para calcular el límite de $f(x)$ cuando $x$ tiende a 0, debemos escribir:

`Limit(f(x), x, 0)`

Lo utilizamos de la siguiente forma:

#  Límites con Manipulación Algebraica

### Factorización y simplificación

$$
\lim_{x \to 2} \frac{x^2 - 4}{x - 2}
$$

**Paso 1:** Sustituimos directamente:

$$
\frac{2^2 - 4}{2 - 2} = \frac{0}{0} \quad \text{(indeterminación)}
$$

**Paso 2:** Factorizamos el numerador:

$$
x^2 - 4 = (x - 2)(x + 2)
$$

**Paso 3:** Simplificamos la fracción:

$$
\frac{(x - 2)(x + 2)}{x - 2} = x + 2, \quad x \neq 2
$$

**Paso 4:** Calculamos el límite sustituyendo ahora:

$$
\lim_{x \to 2} (x + 2) = 2 + 2 = 4
$$



###  Racionalización

$$
\lim_{x \to 0} \frac{\sqrt{x + 1} - 1}{x}
$$

**Paso 1:** Sustituimos directamente:

$$
\frac{\sqrt{0 + 1} - 1}{0} = \frac{1 - 1}{0} = \frac{0}{0} \quad \text{(indeterminación)}
$$

**Paso 2:** Multiplicamos y dividimos por el conjugado para eliminar la raíz:

$$
\frac{\sqrt{x + 1} - 1}{x} \cdot \frac{\sqrt{x + 1} + 1}{\sqrt{x + 1} + 1} = \frac{(x + 1) - 1}{x(\sqrt{x + 1} + 1)} = \frac{x}{x(\sqrt{x + 1} + 1)}
$$

**Paso 3:** Simplificamos el \(x\) en numerador y denominador:

$$
\frac{x}{x(\sqrt{x + 1} + 1)} = \frac{1}{\sqrt{x + 1} + 1}, \quad x \neq 0
$$

**Paso 4:** Sustituimos el límite ahora:

$$
\lim_{x \to 0} \frac{1}{\sqrt{x + 1} + 1} = \frac{1}{\sqrt{1} + 1} = \frac{1}{2}
$$


### 3. Uso de identidades trigonométricas

$$
\lim_{x \to 0} \frac{1 - \cos x}{x^2}
$$

**Paso 1:** Sustituimos directamente:

$$
\frac{1 - \cos 0}{0^2} = \frac{1 - 1}{0} = \frac{0}{0} \quad \text{(indeterminación)}
$$

**Paso 2:** Usamos la identidad trigonométrica:

$$
1 - \cos x = 2 \sin^2\left(\frac{x}{2}\right)
$$

Por lo que:

$$
\lim_{x \to 0} \frac{1 - \cos x}{x^2} = \lim_{x \to 0} \frac{2 \sin^2\left(\frac{x}{2}\right)}{x^2}
$$

**Paso 3:** Reescribimos la expresión para facilitar la evaluación:

$$
= \lim_{x \to 0} 2 \left(\frac{\sin\left(\frac{x}{2}\right)}{x}\right)^2 = \lim_{x \to 0} 2 \left(\frac{\sin\left(\frac{x}{2}\right)}{\frac{x}{2}} \cdot \frac{1}{2}\right)^2
$$

**Paso 4:** Sabemos que

$$
\lim_{t \to 0} \frac{\sin t}{t} = 1,
$$

por lo que:

$$
\lim_{x \to 0} \frac{1 - \cos x}{x^2} = 2 \left(1 \cdot \frac{1}{2}\right)^2 = 2 \left(\frac{1}{2}\right)^2 = 2 \cdot \frac{1}{4} = \frac{1}{2}
$$


###  Manipulación y límite exponencial

$$
\lim_{x \to 0} \frac{e^x - 1}{x}
$$

**Paso 1:** Sustituimos directamente:

$$
\frac{e^0 - 1}{0} = \frac{1 - 1}{0} = \frac{0}{0} \quad \text{(indeterminación)}
$$

**Paso 2:** Reconocemos que esta es una forma clásica, o aplicamos la Ley de L’Hôpital derivando numerador y denominador:

$$
\lim_{x \to 0} \frac{e^x}{1} = e^0 = 1
$$



# Ley de L’Hôpital

La **Ley de L’Hôpital** es una técnica útil para calcular límites que presentan formas indeterminadas del tipo $\frac{0}{0}$ o $\frac{\infty}{\infty}$. Cuando al sustituir directamente el valor al que tiende la variable obtenemos una indeterminación, esta ley nos permite evaluar el límite derivando el numerador y el denominador por separado.

## Enunciado de la Ley

Si

$$
\lim_{x \to a} f(x) = 0 \quad \text{y} \quad \lim_{x \to a} g(x) = 0,
$$

o bien

$$
\lim_{x \to a} f(x) = \pm \infty \quad \text{y} \quad \lim_{x \to a} g(x) = \pm \infty,
$$

y además las derivadas $f'(x)$ y $g'(x)$ existen cerca de $a$ (excepto tal vez en $a$), entonces

$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)},
$$

siempre que el límite del lado derecho exista o sea infinito.



## Procedimiento

1. Identificar que el límite da una forma indeterminada $\frac{0}{0}$ o $\frac{\infty}{\infty}$ al sustituir directamente.
2. Calcular las derivadas de numerador y denominador.
3. Evaluar el límite de la nueva función que es el cociente de las derivadas.
4. Si sigue siendo una forma indeterminada, se puede aplicar la ley de nuevo.



## Ejemplos

### 1. Calcular

$$
\lim_{x \to 0} \frac{\sin x}{x}
$$

**Paso 1:** Sustituimos directamente:

$$
\frac{\sin 0}{0} = \frac{0}{0} \quad \text{indeterminación}
$$

**Paso 2:** Derivamos numerador y denominador:

$$
f'(x) = \cos x, \quad g'(x) = 1
$$

**Paso 3:** Evaluamos el límite del cociente de derivadas:

$$
\lim_{x \to 0} \frac{\cos x}{1} = \cos 0 = 1
$$

### 2. Calcular

$$
\lim_{x \to 0} \frac{e^x - 1}{x}
$$

**Paso 1:** Sustituimos directamente:

$$
\frac{e^0 - 1}{0} = \frac{1 - 1}{0} = \frac{0}{0} \quad \text{indeterminación}
$$

**Paso 2:** Derivamos:

$$
f'(x) = e^x, \quad g'(x) = 1
$$

**Paso 3:** Evaluamos el límite:

$$
\lim_{x \to 0} \frac{e^x}{1} = e^0 = 1
$$

### 3. Calcular

$$
\lim_{x \to 0} \frac{1 - \cos x}{x^2}
$$

**Paso 1:** Sustituimos directamente:

$$
\frac{1 - \cos 0}{0^2} = \frac{0}{0} \quad \text{indeterminación}
$$

**Paso 2:** Derivamos numerador y denominador:

$$
f'(x) = \sin x, \quad g'(x) = 2x
$$

Evaluamos el nuevo límite:

$$
\lim_{x \to 0} \frac{\sin x}{2x} = \frac{0}{0} \quad \text{aún indeterminado}
$$

**Paso 3:** Aplicamos la ley otra vez:

Derivamos de nuevo:

$$
f''(x) = \cos x, \quad g''(x) = 2
$$

Evaluamos:

$$
\lim_{x \to 0} \frac{\cos x}{2} = \frac{1}{2}
$$


## Derivadas

Para poder comprender el concepto de **derivada** primero debemos abordar el problema de la **recta tangente** a una curva. 

La palabra *tangente* se deriva del latín *tangens*, que significa "que toca". Así, una **tangente** a una curva es una línea que toca la curva. En otras palabras, una línea tangente debe tener la misma dirección que la curva en el punto de contacto.

Para un círculo podríamos simplemente seguir la definición de Euclides y decir que la tangente es una línea que cruza el círculo una y sólo una vez (ver *figura a*). Pero para curvas más complicadas esta definición es inadecuada. Por ejemplo, en la *figura b* podemos ver dos líneas $l$ y $t$ que pasan por el punto $P$ en una curva $C$. La línea $l$ cruza a la curva $C$ sólo una vez, pero ciertamente no se parece a lo que pensamos como una tangente. La línea $t$, en cambio, se parece a una tangente pero intercepta a $C$ dos veces.

<img alt="El problema de la tangente" title="El problema de la tangente" src="http://relopezbriega.github.io/images/tangente.png" width="500px" height="400px">

El intento de resolver este problema fue lo que condujo a Fermat a descubrir algunas de las ideas rudimentarias referentes a la noción de **derivada**.

Aunque la derivada se introdujo inicialmente para el estudio del problema de la tangente, pronto se vio que proporcionaba también un instrumento para el cálculo de *velocidades* y, en general, para el estudio de la *variación o tasa de cambio* de una función.

La **derivada** de una función es una medida de la rapidez con la que cambia el valor de dicha función, según cambie el valor de su variable independiente. La derivada de una función es un concepto **local**, es decir, se calcula como el límite de la rapidez de cambio medio de la función en un cierto intervalo, cuando el intervalo considerado para la variable independiente se torna cada vez más pequeño.

Entonces el valor de la derivada de una función en un punto puede interpretarse geométricamente, ya que se corresponde con la **pendiente de la recta tangente** a la gráfica de la función en dicho punto. La recta tangente es a su vez la gráfica de la mejor aproximación lineal de la función alrededor de dicho punto.

Matemáticamente, la derivada es un caso especial de límite. Así, **la derivada de una función $f$ en el punto $a$**, representada por $f'(a)$, es:

$$
f'(a) = \lim_{h \to 0}\frac{f(a + h) - f(a)}{h}
$$

donde $h$ representa la variación de $a$.

Esta misma definición puede ser representada también del siguiente modo, utilizando la notación de Leibniz:

$$
\frac{dy}{dx} = \lim_{dx \to 0}\frac{f(x + dx) - f(x)}{dx}
$$

### Ejemplo:

Si quisiéramos saber cuál es la derivada de la función $f(x) = x^3$, podemos aplicar la definición anterior:

$$
f(x + dx) = (x + dx)^3 = x^3 + 3x^2dx + 3xdx^2 + dx^3
$$

Entonces:

$$
\frac{dy}{dx} = \frac{f(x + dx) - f(x)}{dx} = \frac{x^3 + 3x^2dx + 3xdx^2 + dx^3 - x^3}{dx}
$$

$$
= \frac{3x^2dx + 3xdx^2 + dx^3}{dx} = 3x^2 + 3xdx + dx^2
$$

Cuando $dx \to 0$, los términos con potencias de $dx$ desaparecen:

$$
\frac{d}{dx}x^3 = 3x^2
$$



## Reglas de derivación

Las reglas de derivación nos permiten calcular las derivadas de funciones de forma sistemática sin recurrir a la definición límite cada vez.


### 1. Derivada de función constante

Si $f(x) = c$, donde $c$ es constante,

$$
f'(x) = 0
$$



### 2. Derivada de la función identidad

Si $f(x) = x$,

$$
f'(x) = 1
$$



### 3. Regla de la potencia

Si $f(x) = x^n$, donde $n$ es un número real,

$$
f'(x) = n x^{n-1}
$$



### 4. Derivada de constante por función

Si $f(x) = c \cdot g(x)$, donde $c$ es constante,

$$
f'(x) = c \cdot g'(x)
$$


### 5. Derivada de suma o resta de funciones

Si $f(x) = g(x) \pm h(x)$,

$$
f'(x) = g'(x) \pm h'(x)
$$



### 6. Regla del producto

Para el producto de dos funciones,

$$
\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x)
$$


### 7. Regla del cociente

Para el cociente de dos funciones,

$$
\frac{d}{dx} \left[\frac{f(x)}{g(x)}\right] = \frac{f'(x)g(x) - f(x)g'(x)}{[g(x)]^2}
$$


### 8. Regla de la cadena

Para la composición de funciones $f(g(x))$,

$$
\frac{d}{dx} f(g(x)) = f'(g(x)) \cdot g'(x)
$$


### 9. Derivadas de funciones exponenciales y logarítmicas

- Para $f(x) = e^x$,

$$
f'(x) = e^x
$$

- Para $f(x) = a^x$ (con $a > 0$, $a \neq 1$),

$$
f'(x) = a^x \ln a
$$

- Para $f(x) = \ln x$,

$$
f'(x) = \frac{1}{x}
$$

- Para $f(x) = \log_a x$,

$$
f'(x) = \frac{1}{x \ln a}
$$

# Optimización de una función de una variable con cálculo diferencial

## ¿Qué es la optimización?

La **optimización** consiste en encontrar los valores de entrada (o variables) que minimizan o maximizan una función dada. En el caso de una función $f(x)$ de una sola variable, los extremos (mínimos o máximos locales) ocurren en puntos donde la derivada se anula:

$$
f'(x) = 0
$$

Estos puntos se llaman **puntos críticos**.

## ¿Cómo determinar si es mínimo, máximo o ninguno?

Para clasificar un punto crítico $x = c$, usamos la **segunda derivada** $f''(x)$:

- Si $f''(c) > 0$, entonces $f$ tiene un **mínimo local** en $x = c$.
- Si $f''(c) < 0$, entonces $f$ tiene un **máximo local** en $x = c$.
- Si $f''(c) = 0$, el test es **inconcluso** (puede ser un punto de inflexión o requerir más análisis).



# Optimización de una función de una variable usando cálculo diferencial

## Objetivo
Encontrar los puntos críticos y determinar el mínimo o máximo de la función:

$$
f(x) = x^4 - 4x^3 + 6x^2 + 1
$$

## Paso 1: Derivada primera
Para encontrar los extremos relativos, derivamos la función:

$$
f'(x) = 4x^3 - 12x^2 + 12x
$$

## Paso 2: Igualamos la derivada a cero
Buscamos los puntos críticos:

$$
f'(x) = 4x(x^2 - 3x + 3) = 0
$$

Entonces, los posibles puntos críticos son:

- \( x = 0 \)
- \( x^2 - 3x + 3 = 0 \) → no tiene soluciones reales (discriminante negativo)

Por lo tanto, el único punto crítico real es:

$$
x = 0
$$

## Paso 3: Segunda derivada (criterio de la concavidad)

Calculamos:

$$
f''(x) = 12x^2 - 24x + 12
$$

Evaluamos en \(x = 0\):

$$
f''(0) = 12 > 0
$$

Como \(f''(0) > 0\), entonces \(x = 0\) es **mínimo local**.

## Paso 4: Valor mínimo

$$
f(0) = 0^4 - 4(0)^3 + 6(0)^2 + 1 = 1
$$


- Mínimo local en \(x = 0\), con valor mínimo \(f(0) = 1\)




````python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función
def f(x):
    return x**4 - 4*x**3 + 6*x**2 + 1

# Derivadas para verificar resultados
def f_prime(x):
    return 4*x**3 - 12*x**2 + 12*x

def f_double_prime(x):
    return 12*x**2 - 24*x + 12

# Rango de valores
x_vals = np.linspace(-1, 4, 400)
y_vals = f(x_vals)

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x)$', color='blue')
plt.axvline(x=0, linestyle='--', color='red', label='mínimo local en $x=0$')
plt.scatter(0, f(0), color='red')
plt.title('Optimización de $f(x) = x^4 - 4x^3 + 6x^2 + 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
````

# Cálculo  multivariable

En el cálculo de funciones de varias variables, la derivada se generaliza mediante las **derivadas parciales** y el **gradiente**.

## Derivadas parciales

Si una función depende de varias variables, por ejemplo

$$
f(x, y, z, \ldots),
$$

la derivada parcial con respecto a una variable (por ejemplo, $x$) se obtiene derivando $f$ respecto a $x$ mientras se mantienen las otras variables constantes:

$$
\frac{\partial f}{\partial x} = \lim_{h \to 0} \frac{f(x+h, y, z, \ldots) - f(x, y, z, \ldots)}{h}.
$$

Esta derivada mide la tasa de cambio de la función en la dirección de la variable $x$ mientras las demás permanecen fijas.

## Gradiente

El **gradiente** de una función escalar $f:\mathbb{R}^n \to \mathbb{R}$ es un vector que agrupa todas las derivadas parciales:

$$
\nabla f = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \ldots, \frac{\partial f}{\partial x_n} \right).
$$

- **Interpretación geométrica:** El gradiente apunta en la dirección de máxima tasa de crecimiento de la función.
- La magnitud del gradiente indica la rapidez con la que la función crece en esa dirección.
- Por ejemplo, en una función de dos variables $f(x,y)$, el gradiente es un vector en el plano $xy$ que señala hacia donde la función aumenta más rápido.

## Derivada total y diferenciabilidad

La derivada total en un punto combina las derivadas parciales para aproximar el cambio en la función cuando todas las variables cambian simultáneamente:

$$
df = \frac{\partial f}{\partial x} dx + \frac{\partial f}{\partial y} dy + \cdots
$$

Si existe un plano tangente que aproxima la función cerca del punto, decimos que $f$ es diferenciable en ese punto.


## Ejemplos

1. Función

$$
f(x,y) = x^2 y + 3xy^3.
$$

Derivadas parciales:

$$
\frac{\partial f}{\partial x} = 2xy + 3y^3,
$$

$$
\frac{\partial f}{\partial y} = x^2 + 9xy^2.
$$

Gradiente:

$$
\nabla f = \left( 2xy + 3y^3, \quad x^2 + 9xy^2 \right).
$$

Por ejemplo, en el punto $(1,2)$:

$$
\nabla f(1,2) = \left( 2 \cdot 1 \cdot 2 + 3 \cdot 2^3, \quad 1^2 + 9 \cdot 1 \cdot 2^2 \right) = (4 + 24, \, 1 + 36) = (28, 37).
$$



Consideremos la función
$$
f(x,y) = x^2 + y^2.
$$
Esta función es **cóncava hacia arriba** porque su Hessiana es positiva definida en todo $\mathbb{R}^2$.


### Punto crítico y mínimo

El punto $(0,0)$ es un punto crítico, ya que
$$
\nabla f(0,0) = \left(\frac{\partial f}{\partial x}(0,0), \frac{\partial f}{\partial y}(0,0)\right) = (0,0).
$$
Además, dado que la función es cóncava hacia arriba, $(0,0)$ es un **mínimo local** global.



###  Plano tangente en $(x_0, y_0)$

El plano tangente a $f$ en un punto $(x_0,y_0)$ está dado por:
$$
z = f(x_0,y_0) + \frac{\partial f}{\partial x}(x_0,y_0)(x - x_0) + \frac{\partial f}{\partial y}(x_0,y_0)(y - y_0).
$$

En $(0,0)$, esto es:
$$
z = f(0,0) + 0 \cdot (x - 0) + 0 \cdot (y - 0) = 0.
$$


### Derivadas parciales

$$
\frac{\partial f}{\partial x} = 2x, \quad \frac{\partial f}{\partial y} = 2y.
$$


###  Gradiente

El gradiente de $f$ es el vector:
$$
\nabla f(x,y) = \left(2x, 2y\right).
$$
Este vector apunta en la dirección de máxima tasa de aumento de la función, y su magnitud indica qué tan rápido cambia $f$ en esa dirección.



### Derivada direccional

La derivada direccional de $f$ en un punto $\mathbf{a}=(x_0,y_0)$ en la dirección de un vector unitario $\mathbf{u} = (u_1, u_2)$ es
$$
D_{\mathbf{u}} f(\mathbf{a}) = \nabla f(\mathbf{a}) \cdot \mathbf{u} = \frac{\partial f}{\partial x}(x_0,y_0) u_1 + \frac{\partial f}{\partial y}(x_0,y_0) u_2.
$$

Representa la tasa de cambio de la función en la dirección $\mathbf{u}$.



###  Derivada total

La derivada total de $f$ en $\mathbf{a}$ es una función lineal que aproxima el cambio en $f$ cerca de $\mathbf{a}$:
$$
df = \frac{\partial f}{\partial x} dx + \frac{\partial f}{\partial y} dy.
$$

Es la mejor aproximación lineal a $f$ en un entorno pequeño alrededor de $\mathbf{a}$.


- El gradiente indica la dirección y magnitud del cambio más rápido de la función.
- La derivada direccional mide el cambio en cualquier dirección específica.
- El plano tangente es la superficie lineal que toca a $f$ en un punto y aproxima $f$ localmente.
- La derivada total es la generalización de las derivadas parciales, representando el cambio total infinitesimal de la función.
- El punto $(0,0)$ es el mínimo global, donde el gradiente es cero y el plano tangente coincide con el plano $z=0$.


````python
import numpy as np
import matplotlib.pyplot as plt

# Función
def f(x, y):
    return x**2 + y**2

# Gradiente (derivadas parciales)
def dfdx(x, y):
    return 2*x

def dfdy(x, y):
    return 2*y

# Punto de interés
x0, y0 = 0, 0
z0 = f(x0, y0)

# Plano tangente en (x0, y0)
def plano_tangente(x, y):
    return z0 + dfdx(x0, y0)*(x - x0) + dfdy(x0, y0)*(y - y0)

# Grid
x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
Z_tangent = plano_tangente(X, Y)

# Gradiente en (x0, y0)
grad_x = dfdx(x0, y0)
grad_y = dfdy(x0, y0)

fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111, projection='3d')

# Superficie de la función
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Plano tangente
ax.plot_surface(X, Y, Z_tangent, color='orange', alpha=0.5)

# Punto mínimo
ax.scatter(x0, y0, z0, color='red', s=60, label='Mínimo (0,0)')

# Vector gradiente (escalado para visibilidad)
# El vector gradiente está en el plano XY, su componente Z la tomamos igual a 0
scale = 0.5  # factor para que se vea el vector aunque sea pequeño
ax.quiver(x0, y0, z0, grad_x, grad_y, 0, length=scale, color='blue', label='Gradiente en (0,0)')

ax.set_title('Función $f(x,y) = x^2 + y^2$ con plano tangente y vector gradiente')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

plt.show()
````

### ¿Qué representa el campo de gradientes?

El **campo de gradientes** de una función $f(x,y)$ es un campo vectorial que asigna a cada punto $(x,y)$ un vector que apunta en la dirección de la máxima tasa de incremento de la función en ese punto. La magnitud de este vector indica qué tan rápido crece la función en esa dirección.

Para nuestra función
$$
f(x,y) = x^2 + y^2,
$$
el gradiente es
$$
\nabla f(x,y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (2x, 2y).
$$

Este vector apunta "hacia afuera" desde el origen y su longitud aumenta conforme nos alejamos del mínimo en $(0,0)$.



### Derivadas direccionales

La derivada direccional de $f$ en un punto $(x_0,y_0)$ en la dirección del vector unitario $\mathbf{u} = (u_1,u_2)$ se define como:
$$
D_{\mathbf{u}} f(x_0,y_0) = \nabla f(x_0,y_0) \cdot \mathbf{u} = \frac{\partial f}{\partial x}(x_0,y_0) u_1 + \frac{\partial f}{\partial y}(x_0,y_0) u_2.
$$
sto representa la tasa de cambio de $f$ en la dirección $\mathbf{u}$.



### Ejemplo de derivadas direccionales en $(0.5,0.5)$

$$
\mathbf{u} = \frac{1}{\sqrt{2}}(1,1)
$$
(normalizada para ser unidad).

Calculamos:
$$
\nabla f(0.5,0.5) = (1,1),
$$
por lo tanto:
$$
D_{\mathbf{u}} f(0.5,0.5) = (1,1) \cdot \frac{1}{\sqrt{2}}(1,1) = \frac{1}{\sqrt{2}} (1 + 1) = \frac{2}{\sqrt{2}} = \sqrt{2} \approx 1.414.

Esto indica función crece con tasa aproximadamente 1.414 en esa dirección en el punto $(0.5,0.5)$.

### Matriz Hessiana

La **matriz Hessiana** de una función escalar $f(x_1, x_2, \dots, x_n)$ es una matriz cuadrada que contiene todas las **segundas derivadas parciales** de $f$.

Se denota como:

$$
H_f(x) =
\begin{bmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n} \\\\
\frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} & \cdots & \frac{\partial^2 f}{\partial x_2 \partial x_n} \\\\
\vdots & \vdots & \ddots & \vdots \\\\
\frac{\partial^2 f}{\partial x_n \partial x_1} & \frac{\partial^2 f}{\partial x_n \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_n^2}
\end{bmatrix}
$$


La Hessiana nos **describe la curvatura local** de la función multivariable.

- Si la matriz Hessiana es **positiva definida** en un punto crítico, entonces $f$ tiene un **mínimo local** ahí.
- Si es **negativa definida**, $f$ tiene un **máximo local**.
- Si tiene **valores propios positivos y negativos**, el punto es un **punto de silla** (ni mínimo ni máximo).



###  Ejemplo

Para la función $f(x, y) = x^2 + y^2$, tenemos:

$$
\frac{\partial^2 f}{\partial x^2} = 2,\quad
\frac{\partial^2 f}{\partial y^2} = 2,\quad
\frac{\partial^2 f}{\partial x \partial y} = 0
$$

Entonces la matriz Hessiana es:

$$
H_f(x, y) =
\begin{bmatrix}
2 & 0 \\\\
0 & 2
\end{bmatrix}
$$

Esta matriz es **positiva definida**, por lo que $f$ tiene un **mínimo local** (en este caso, también global) en su punto crítico.

# Teorema de Taylor: Una y Varias Variables

El **Teorema de Taylor** permite aproximar una función suave mediante un polinomio construido a partir de sus derivadas en un punto. Es una herramienta fundamental en análisis y aplicaciones numéricas.

## Taylor en Una Variable

Sea $f : \mathbb{R} \to \mathbb{R}$ una función $n$ veces derivable en un punto $a \in \mathbb{R}$. El polinomio de Taylor de orden $n$ centrado en $a$ es:

$$
P_n(x) = f(a) + f'(a)(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \cdots + \frac{f^{(n)}(a)}{n!}(x-a)^n
$$

La forma general es:

$$
P_n(x) = \sum_{k=0}^{n} \frac{f^{(k)}(a)}{k!}(x-a)^k
$$

El error o residuo está dado por el **término de Lagrange**:

$$
R_n(x) = \frac{f^{(n+1)}(\xi)}{(n+1)!}(x - a)^{n+1}, \quad \text{para algún } \xi \in (a, x)
$$

Este polinomio aproxima a $f(x)$ cerca del punto $x = a$.

### Ejemplo

Para $f(x) = \cos x$, el polinomio de Taylor de orden 4 centrado en $a = 0$ es:

$$
P_4(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!}
$$

## Taylor en Varias Variables

Sea ahora $f : \mathbb{R}^n \to \mathbb{R}$ una función suficientemente diferenciable. El polinomio de Taylor de segundo orden centrado en $\mathbf{a} \in \mathbb{R}^n$ es:

$$
f(\mathbf{x}) \approx f(\mathbf{a}) + \nabla f(\mathbf{a}) \cdot (\mathbf{x} - \mathbf{a}) + \frac{1}{2} (\mathbf{x} - \mathbf{a})^T H_f(\mathbf{a}) (\mathbf{x} - \mathbf{a})
$$

donde:

- $\nabla f(\mathbf{a})$ es el **gradiente** de $f$ en $\mathbf{a}$
- $H_f(\mathbf{a})$ es la **matriz Hessiana** de $f$ en $\mathbf{a}$

En componentes:

$$
f(\mathbf{x}) \approx f(\mathbf{a}) + \sum_{i=1}^n \frac{\partial f}{\partial x_i}(\mathbf{a})(x_i - a_i) + \frac{1}{2} \sum_{i,j=1}^n \frac{\partial^2 f}{\partial x_i \partial x_j}(\mathbf{a})(x_i - a_i)(x_j - a_j)
$$

Este polinomio se utiliza para aproximar funciones multivariables cerca del punto $\mathbf{a}$ y es útil, por ejemplo, para encontrar máximos, mínimos y puntos de silla.



### Ejemplo en dos variables

Sea:

$$
f(x, y) = x^2 + y^2
$$

Queremos hallar el polinomio de Taylor de orden 2 centrado en $(a, b) = (0, 0)$.

 **Evaluamos la función en el punto**:

$$
f(0, 0) = 0
$$

 **Gradiente**:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (2x, 2y)
$$

Entonces:

$$
\nabla f(0, 0) = (0, 0)
$$

 **Hessiana**:

$$
H_f(x, y) =
\begin{bmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{bmatrix}
=
\begin{bmatrix}
2 & 0 \\
0 & 2
\end{bmatrix}
$$

Entonces:

$$
H_f(0,0) = 
\begin{bmatrix}
2 & 0 \\
0 & 2
\end{bmatrix}
$$

 **Polinomio de Taylor de segundo orden**:

Ya que el gradiente es cero, solo queda el término cuadrático:

$$
f(x, y) \approx 0 + \frac{1}{2}
\begin{bmatrix}
x & y
\end{bmatrix}
\begin{bmatrix}
2 & 0 \\
0 & 2
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
= x^2 + y^2
$$

Como la función ya es un polinomio de segundo orden, ¡el desarrollo de Taylor es exacto!




### Ejemplo en dos variables

Sea:

$$
f(x, y) = \sin(xy)
$$

Queremos hallar el polinomio de Taylor de orden 2 centrado en $(a, b) = (0, 0)$.



**Evaluamos la función en el punto**:

$$
f(0, 0) = \sin(0 \cdot 0) = 0
$$



**Gradiente**:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
= \left( y \cos(xy),\ x \cos(xy) \right)
$$

Entonces:

$$
\nabla f(0, 0) = (0, 0)
$$


**Hessiana**:

Calculamos las segundas derivadas parciales:

- $\displaystyle \frac{\partial^2 f}{\partial x^2} = -y^2 \sin(xy)$  
- $\displaystyle \frac{\partial^2 f}{\partial y^2} = -x^2 \sin(xy)$  
- $\displaystyle \frac{\partial^2 f}{\partial x \partial y} = \cos(xy) - xy \sin(xy)$

Entonces:

$$
H_f(x, y) =
\begin{bmatrix}
- y^2 \sin(xy) & \cos(xy) - xy \sin(xy) \\\\
\cos(xy) - xy \sin(xy) & - x^2 \sin(xy)
\end{bmatrix}
$$

Evaluamos en $(0, 0)$:

- $\displaystyle \frac{\partial^2 f}{\partial x^2}(0,0) = 0$
- $\displaystyle \frac{\partial^2 f}{\partial y^2}(0,0) = 0$
- $\displaystyle \frac{\partial^2 f}{\partial x \partial y}(0,0) = 1$

Entonces:

$$
H_f(0,0) =
\begin{bmatrix}
0 & 1 \\\\
1 & 0
\end{bmatrix}
$$


**Polinomio de Taylor de segundo orden**:

Ya que $f(0,0) = 0$ y el gradiente también es cero, solo queda el término cuadrático:

$$
f(x, y) \approx \frac{1}{2}
\begin{bmatrix}
x & y
\end{bmatrix}
\begin{bmatrix}
0 & 1 \\\\
1 & 0
\end{bmatrix}
\begin{bmatrix}
x \\\\
y
\end{bmatrix}
= \frac{1}{2} (2xy) = xy
$$



El desarrollo de Taylor de segundo orden de $f(x, y) = \sin(xy)$ alrededor de $(0,0)$ es:

$$
f(x, y) \approx xy
$$

Este polinomio es una buena aproximación de $\sin(xy)$ cerca del origen.


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Función original
def f(x, y):
    return np.sin(x * y)

# Aproximación de Taylor de segundo orden en (0,0)
def taylor_orden2(x, y):
    return x * y

# Rango de valores
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)
X, Y = np.meshgrid(x, y)

# Evaluamos ambas funciones
Z_f = f(X, Y)
Z_taylor = taylor_orden2(X, Y)

# Crear figuras
fig = plt.figure(figsize=(14, 6))

# Gráfica de la función original
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z_f, cmap='viridis', alpha=0.9)
ax1.set_title(r'$f(x, y) = \sin(xy)$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')

# Gráfica de la aproximación de Taylor
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z_taylor, cmap='plasma', alpha=0.9)
ax2.set_title(r'Taylor orden 2: $f(x, y) \approx xy$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('Aproximación')

plt.tight_layout()
plt.show()
````

# Error  en el Teorema de Taylor

Cuando usamos el **Teorema de Taylor** para aproximar funciones, existen dos tipos principales de errores que debemos considerar:

## Error de truncamiento

Este error proviene de truncar la serie de Taylor después de un número finito de términos. Es el error **teórico** que mide la diferencia entre la función real y el polinomio de Taylor.

Por ejemplo, para una función de una variable:

$$
f(x) = P_n(x) + R_n(x)
$$

donde

$$
R_n(x) = O\left((x - a)^{n+1}\right)
$$

es el error de truncamiento (término restante).

## Error de redondeo

El **error de redondeo** es un error **numérico** que surge al realizar cálculos en una computadora debido a la representación finita de los números reales (precisión limitada de coma flotante).

- Los números reales se representan con una cantidad limitada de dígitos (por ejemplo, 16 dígitos en doble precisión).
- Las operaciones aritméticas (suma, multiplicación, etc.) no son exactas, se aproximan.
- Al calcular derivadas, valores de funciones y evaluaciones de polinomios, se acumulan pequeñas imprecisiones.

## Notación del error de redondeo en la aproximación de Taylor

### Una variable

Al aproximar $f(x)$ alrededor de $a$ con un polinomio de orden $n$:

$$
f(x) = P_n(x) + O\left((x - a)^{n+1}\right)
$$

La evaluación numérica de $P_n(x)$ introduce un error de redondeo, que suele ser proporcional a la precisión de la máquina (por ejemplo, $\varepsilon \approx 10^{-16}$) y se puede representar como:

$$
\text{Error total} = O\left((x - a)^{n+1}\right) + O(\varepsilon)
$$

### Varias variables

Para una función $f(\mathbf{x})$, con $\mathbf{x} \in \mathbb{R}^m$, el polinomio de Taylor de orden 2 alrededor de $\mathbf{a}$ es:

$$
f(\mathbf{x}) = f(\mathbf{a}) + \nabla f(\mathbf{a}) \cdot (\mathbf{x} - \mathbf{a}) + \frac{1}{2} (\mathbf{x} - \mathbf{a})^T H_f(\mathbf{a}) (\mathbf{x} - \mathbf{a}) + O(\|\mathbf{x} - \mathbf{a}\|^3)
$$

El error de truncamiento es del orden:

$$
O(\|\mathbf{x} - \mathbf{a}\|^{3})
$$

y el error de redondeo numérico acumulado durante la evaluación del polinomio puede expresarse como:

$$
O(\varepsilon)
$$

donde $\varepsilon$ es la precisión de la máquina.



La aproximación final numérica satisface:

$$
\boxed{
f(\mathbf{x}) \approx P_n(\mathbf{x}) + O\left(\|\mathbf{x} - \mathbf{a}\|^{n+1}\right) + O(\varepsilon)
}
$$

donde

- $O\left(\|\mathbf{x} - \mathbf{a}\|^{n+1}\right)$ es el error de truncamiento.
- $O(\varepsilon)$ es el error de redondeo por cálculos numéricos finitos.

Ambos errores determinan la precisión de la aproximación de Taylor en computación.
