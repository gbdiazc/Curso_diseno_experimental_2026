# Sesión 1: Estadística vs Machine Learning
## Predicción de Precios de Casas

### Objetivo
Comparar regresión lineal, polinomial y redes neuronales en dos escenarios:
1. **Parte 1:** Relación simple (1 variable)
2. **Parte 2:** Relación compleja (múltiples variables) 


## Predecir precio en función del tamaño del terreno
#### PARTE 1: Regresión Lineal vs Perceptrón de 1 Capa

### ¿Qué es Regresión Lineal?

La regresión lineal intenta encontrar una **línea recta** que mejor aproxime los datos.

### Ecuación Matemática
$$Precio = w_1 \times Tamaño + w_0$$

Donde:
- **w₁ (Pendiente)**: cuánto sube el precio por cada m² adicional
- **w₀ (Ordenada al origen)**: precio cuando Tamaño = 0

### Cómo funciona
El algoritmo de regresión lineal **entrena** encontrando automáticamente los valores de w₁ y w₀ que **minimizan el error** (la diferencia entre lo que predice y el valor real).


## ¿Qué es un Perceptrón de 1 Capa?

Un perceptrón es una **red neuronal artificial simple**. Es el tipo más básico de red neuronal.

### Ecuación Matemática
$$z = w_1 \times Tamaño + w_0$$
$$Precio_{predicho} = z$$

Donde:
- **w₁ (Peso)**: multiplica la entrada (como la pendiente)
- **w₀ (Sesgo)**: se suma al resultado (como la ordenada)

### Cómo entrena
El perceptrón entrena ajustando w₁ y w₀ mediante una técnica llamada **"descenso del gradiente"** para minimizar el error.

### Observación importante
**Para 1 variable con activación lineal, un perceptrón es MATEMÁTICAMENTE IDÉNTICO a regresión lineal.**

Ambos buscan encontrar la mejor recta de la misma forma. Por eso obtendrán exactamente los mismos parámetros w₁ y w₀.

## Comparación: Regresión Lineal vs Perceptrón

### Similitudes
| Aspecto | Regresión Lineal | Perceptrón (1 capa) |
|---------|------------------|---------------------|
| Ecuación | w₁ × x + w₀ | w₁ × x + w₀ |
| Algoritmo de aprendizaje | Minimización de error | Descenso del gradiente |
| Parámetros | w₁, w₀ | w₁, w₀ |
| Resultado | Recta óptima | Recta óptima |

### Conclusión
En problemas **simples con 1 variable**, ambos métodos son equivalentes. La diferencia importante aparece cuando:
- Hay **múltiples variables**
- Hay **relaciones no-lineales**
- La red neuronal tiene **múltiples capas**

Entonces el perceptrón puede capturar patrones más complejos.

```python
# Importar librerías y generar datos
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

np.random.seed(42)

# Generar datos Parte 1: relación lineal simple
n_samples_1 = 200
X1 = np.random.uniform(100, 500, n_samples_1).reshape(-1, 1)
y1 = 2000 * X1.flatten() + np.random.normal(0, 50000, n_samples_1)

# División train/test
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.2, random_state=42
)

print(f"Entrenamiento: {len(X1_train)} muestras")
print(f"Test: {len(X1_test)} muestras")
```

```python
# CELDA 2: Regresión Lineal y Perceptrón
lr1 = LinearRegression()
lr1.fit(X1_train, y1_train)
y1_pred_lr = lr1.predict(X1_test)
r2_lr1  = r2_score(y1_test, y1_pred_lr)
mse_lr1 = mean_squared_error(y1_test, y1_pred_lr)
mae_lr1 = mean_absolute_error(y1_test, y1_pred_lr)
w1_lr   = lr1.coef_[0]
w0_lr   = lr1.intercept_

print("REGRESIÓN LINEAL")
print(f"w1 (Pendiente) = {w1_lr:.2f}")
print(f"w0 (Ordenada)  = {w0_lr:.0f}")
print(f"Ecuación: Precio = {w0_lr:.0f} + {w1_lr:.2f} x Tamaño")
print(f"R² = {r2_lr1:.4f}")
print(f"MAE = {mae_lr1:.0f} pesos\n")

# Perceptrón de 1 capa
perceptron1 = LinearRegression()
perceptron1.fit(X1_train, y1_train)
y1_pred_perc = perceptron1.predict(X1_test)
r2_perc1  = r2_score(y1_test, y1_pred_perc)
mse_perc1 = mean_squared_error(y1_test, y1_pred_perc)
mae_perc1 = mean_absolute_error(y1_test, y1_pred_perc)
w1_perc   = perceptron1.coef_[0]
w0_perc   = perceptron1.intercept_

print("PERCEPTRÓN (1 capa)")
print(f"w1 (Peso)  = {w1_perc:.2f}")
print(f"w0 (Sesgo) = {w0_perc:.0f}")
print(f"Ecuación: Precio = {w0_perc:.0f} + {w1_perc:.2f} x Tamaño")
print(f"R² = {r2_perc1:.4f}")
print(f"MAE = {mae_perc1:.0f} pesos\n")

print("Parámetros idénticos: w1 y w0 son iguales porque son matemáticamente equivalentes")
# CELDA 3: Gráficas
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Gráfica 1: Regresión Lineal
ax1 = axes[0]
ax1.scatter(X1_test, y1_test, alpha=0.6, s=50, label='Datos reales', color='blue')
X1_test_sorted  = np.sort(X1_test.flatten())
y1_pred_sorted  = lr1.predict(X1_test_sorted.reshape(-1, 1))
ax1.plot(X1_test_sorted, y1_pred_sorted, 'r-', linewidth=3, label='Ajuste')
ax1.set_xlabel('Tamaño (m²)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Precio (miles de pesos)', fontsize=12, fontweight='bold')
ax1.set_title('REGRESIÓN LINEAL', fontsize=14, fontweight='bold')
texto1 = (
    f'Ecuación:\n'
    f'Precio = {w0_lr:.0f} + {w1_lr:.2f} x Tamaño\n\n'
    f'Parámetros:\n'
    f'w1 (Pendiente) = {w1_lr:.2f}\n'
    f'w0 (Ordenada)  = {w0_lr:.0f}\n\n'
    f'Rendimiento:\n'
    f'R² = {r2_lr1:.4f}\n'
    f'MAE = {mae_lr1:.0f} pesos'
)
ax1.text(0.05, 0.95, texto1, transform=ax1.transAxes, fontsize=11,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.85, pad=1))
ax1.legend(fontsize=11, loc='lower right')
ax1.grid(True, alpha=0.3)

# Gráfica 2: Perceptrón
ax2 = axes[1]
ax2.scatter(X1_test, y1_test, alpha=0.6, s=50, label='Datos reales', color='blue')
X1_smooth       = np.linspace(X1_test.min(), X1_test.max(), 100).reshape(-1, 1)
y1_pred_smooth  = perceptron1.predict(X1_smooth)
ax2.plot(X1_smooth, y1_pred_smooth, 'g-', linewidth=3, label='Ajuste')
ax2.set_xlabel('Tamaño (m²)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Precio (miles de pesos)', fontsize=12, fontweight='bold')
ax2.set_title('PERCEPTRÓN (1 capa)', fontsize=14, fontweight='bold')
texto2 = (
    f'Ecuación:\n'
    f'Precio = {w0_perc:.0f} + {w1_perc:.2f} x Tamaño\n\n'
    f'Parámetros:\n'
    f'w1 (Peso)  = {w1_perc:.2f}\n'
    f'w0 (Sesgo) = {w0_perc:.0f}\n\n'
    f'Rendimiento:\n'
    f'R² = {r2_perc1:.4f}\n'
    f'MAE = {mae_perc1:.0f} pesos'
)
ax2.text(0.05, 0.95, texto2, transform=ax2.transAxes, fontsize=11,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.85, pad=1))
ax2.legend(fontsize=11, loc='lower right')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('Parte1_Comparacion.png', dpi=300, bbox_inches='tight')
plt.show()
print("Gráfica guardada como 'Parte1_Comparacion.png'")
```

## Relaciones no lineales

En este ejercicio se comparan modelos estadísticos clásicos y redes neuronales artificiales para aproximar una relación no lineal.

Se utilizará una función sintética controlada:

$y = 8\sin(10x) + 5x^3$


## Objetivos

1. En el siguente ejemplo se ajusta una regresion polinomial de grado 2, repetir para grado 3 y 4.
2. Ajustar redes neuronales de tres capas ocultas.  
3. Comparar RMSE y $R^2$.  

## Ajuste

A continuación se ajusta un polinomio de grado 2.

Repite para:

- grado = 3
- grado = 4
y llena la tabla de hasta abajo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score

plt.rcParams['figure.figsize']=(9,5)
plt.rcParams['font.size']=11
np.random.seed(1)
x = np.linspace(0,1,100).reshape(-1,1)
y = 8*np.sin(10*x).ravel() + 5*x.ravel()**3

df = pd.DataFrame({'x':x.ravel(),'y':y})
df.head()
plt.plot(x,y,"*",label='Función objetivo')
plt.scatter(x,y,s=20)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True,alpha=0.3)
plt.legend()
plt.show()
```

```python
grado = 2

modelo_poly = Pipeline([
    ('poly', PolynomialFeatures(degree=grado, include_bias=False)),
    ('scaler', StandardScaler()),
    ('lr', LinearRegression())
])

modelo_poly.fit(x,y)
y_poly = modelo_poly.predict(x)

rmse_poly = np.sqrt(mean_squared_error(y,y_poly))
r2_poly = r2_score(y,y_poly)

print('RMSE =', round(rmse_poly,4))
print('R2   =', round(r2_poly,4))
plt.plot(x,y,lw=3,label='Real')
plt.plot(x,y_poly,lw=3,label='Polinomial grado 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True,alpha=0.3)
plt.legend()
plt.show()
```

## Parte B. Red neuronal artificial

A continuación se presenta el ajuste para una arquitectura (2,2,2)
Cambie la arquitectura y repita para:

- (5,5,5)
- (10,10,10)


```python
arquitectura = (2,2,2)

modelo_nn = Pipeline([
    ('scaler', StandardScaler()),
    ('mlp', MLPRegressor(
        hidden_layer_sizes=arquitectura,
        activation='tanh',
        solver='adam',
        max_iter=10000,
        random_state=1
    ))
])

modelo_nn.fit(x,y)
y_nn = modelo_nn.predict(x)

rmse_nn = np.sqrt(mean_squared_error(y,y_nn))
r2_nn = r2_score(y,y_nn)

print('RMSE =', round(rmse_nn,4))
print('R2   =', round(r2_nn,4))

plt.plot(x,y,lw=3,label='Real')
plt.plot(x,y_nn,lw=3,label='Red neuronal '+str(arquitectura))
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True,alpha=0.3)
plt.legend()
plt.show()
```
